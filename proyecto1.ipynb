Ejercicio N°1:
Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

```python
import pandas as pd
```

```python
ventas = {}
descuentos = {}
ventafinal = {}
dias = {}
totaldias = 5
for i in range(totaldias):
    dias[i] = i+1
    ventas[i] = float(input('Introduce las ventas del dia ' + str(i+1) +': '))
    descuentos[i] = ventas[i] * 0.05
    ventafinal[i] = ventas[i] - descuentos[i]
items = {'Dia': dias, 'Venta': ventas, 'Descuento': descuentos, 'Venta final': ventafinal}
pdventas = pd.DataFrame(items)
pdventas
```

Ejercicio No 2
Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.El resultado final debe ser una serie con los estadísticos.

```python
def ingresos_mensuales(sueldo):
    sueldo = pd.Series(sueldo)
    calculo = pd.Series([sueldo.min(), sueldo.max(), sueldo.mean(), sueldo.std()], index=['Min', 'Max', 'Media', 'Desviación típica'])
    return calculo

sueldo = {'Tranajador 1':1500, 'Tranajador 2':1500, 'Tranajador 3':2250, 'Tranajador 4': 1800, 'Tranajador 5': 2200}
ingresos_mensuales(sueldo)
```

Ejercicio No 3
Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla:

    Mes     Ventas    Gastos
    Enero   30500    22000
    Febrero 35600    23450
    Marzo   28300    18100
    Abril   33900    35700
    Mayo    42500    32450

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'], 
         'Ventas':[30500, 35600, 28300, 33900, 42500], 
         'Gastos':[22000, 23400, 18100, 20700, 50000]}
tabla = pd.DataFrame(datos)
tabla
```

```python
#1. Columna 1 = Gasto/Ventas*100.
columna1 = round(tabla['Gastos']/tabla['Ventas']*100, 2)
tabla['Columna1']=columna1
#2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,caso contrario imprimir “Meta no Superada”.
tabla['Columna2'] = tabla['Ventas'].apply(lambda x:
                                            "Meta Superada" if (x >= 30000) else "Meta no Superada"
                                        )
#3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
tabla['Columna3'] = tabla['Gastos'] - tabla['Ventas']
tabla['Columna3'] = tabla['Columna3'].apply(lambda x:
                                            "0" if (x < 0) else "Sobre gasto de "+str(x)
                                        )
tabla
```

```python
#4. Obtener el total de las ventas de los últimos 3 meses
import numpy as np
dfVentas = tabla['Ventas'][-3:]
totalVenta = np.sum(dfVentas, axis=0)
totalVenta
```

```python
#5. Obtener el total de gastos de febrero a abril
dfGastos = tabla['Gastos'][1:4]
totalGasto = np.sum(dfGastos, axis=0)
totalGasto
```
